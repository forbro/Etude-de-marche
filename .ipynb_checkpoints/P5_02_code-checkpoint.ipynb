```python
#Préparation des données 
n_comp = 4
X = df_pca.values
names = df.set_index('pays').index
features = df_pca.columns

# Centrage et Réduction
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)

# Clustering hiérarchique
Z = linkage(X_scaled, 'ward')

# Affichage du dendrogramme
plt.figure(figsize=(10,25))
plot_dendrogram(Z, names);
plt.show();
```

### Coupage du dendrogramme en 5 clusters

```python
#Coupage du dendrogramme en 5 clusters
clusters = fcluster(Z, 5, criterion='maxclust')
#clusters = fcluster(Z, 159, criterion='distance') # ligne équivalente à la précédente

# Comparaison des clusters trouvés avec les classification des pays
pays = pd.DataFrame({"pays": names, "cluster": clusters})
pays.to_csv('Liste_pays.csv', index=False)
```

```python
# Merger le numéro du cluster au pays correspondant
df = pd.merge(df, pays, on='pays') 
illustrative_var = df['cluster']
ddf = df[['pays', 'taux_daccroiss_pop', 'dispo_alim_kcal_p_an_ani','nbre_habitant', 'proport_prot_ani','pib_habitant']]
ddf = ddf.set_index('pays')
```

```python
from sklearn.cluster import KMeans
n_clust = 5
# Clustering par K-means
km = KMeans(n_clusters=n_clust).fit(X_scaled)

# Récupération des clusters attribués à chaque individu
clusters = km.labels_

# Affichage du clustering par projection des individus sur le premier plan factoriel
pca = decomposition.PCA(n_components=n_comp).fit(X_scaled)
X_projected = pca.transform(X_scaled)

# Projection des individus
display_factorial_planes(X_projected, n_comp, pca, [(0,1)], illustrative_var = illustrative_var)
```

```python
data = df.copy()
data.drop(['code','num'], axis=1, inplace=True)
```

```python
def scaler(data):
    data_num=data.select_dtypes(exclude='object')
    for col in list(data_num.columns):
        data[col] = (data[col]-data[col].min())/(data[col].max()-data[col].min())
    return data
```

```python
data = scaler(data)
```

```python
scoring = df.copy()

#Score 
scoring['score_1'] = (data['taux_daccroiss_pop']*1+1
                      -data['dispo_alim_kcal_p_an_ani']
                      +data['nbre_habitant']*1+1
                      -data['proport_prot_ani'])/4
```

```python
df_quantiles = df.copy()
df_quantiles = df_quantiles.quantile(q=[0.25,0.5,0.75])
df_quantiles = df_quantiles.to_dict()
```

```python
def benefique(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
def non_benefique(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
scoring['t_dac'] = scoring['taux_daccroiss_pop'].apply(benefique, args=('taux_daccroiss_pop',df_quantiles,))
scoring['kcal'] = scoring['dispo_alim_kcal_p_an_ani'].apply(non_benefique, args=('dispo_alim_kcal_p_an_ani',df_quantiles,))
scoring['pop'] = scoring['nbre_habitant'].apply(benefique, args=('nbre_habitant',df_quantiles,))
scoring['proport'] = scoring['proport_prot_ani'].apply(non_benefique, args=('proport_prot_ani',df_quantiles,))
#rajoute les autres aussi
```

```python
scoring['score'] = (scoring['t_dac']+ scoring['kcal']+scoring['pop']+scoring['proport'])/16
```

```python
print(scoring[['score','score_1']].corr())
#pour quantifier la dependance

sns.lmplot(x="score", y="score_1", data=scoring);
#pour visualiser
```

```python
from scipy import stats
numeric_columns = df.select_dtypes(include = ['int32','float64']).columns
#numeric_columns = numeric_columns[1:]
for column in numeric_columns:
    print('________\n{}'.format(column))
    k2, p = stats.normaltest(df[column], 
                             axis=0,
                             nan_policy = 'omit')
    alpha = 5e-2
    print("p = {:g}".format(p))
    if p < alpha:  # null hypothese: x a une distribution normale
        print("H0 est rejetée : {} n'est pas de distribution normale".format(column))
    else:
        print("H0 ne peut être rejetée :{}, on considère l'hypothèse de normalité".format(column))
```

### Test de comparaison

```python
#H0: "le taux d'accroissement et les clusters sont INDEPENDANTS"
# get ANOVA table as R like output
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Ordinary Least Squares (OLS) model
model = ols('taux_daccroiss_pop ~ cluster', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table

#si p-value(PR(>F))<0.05 alors il y a dependance entre les deux variables

```
